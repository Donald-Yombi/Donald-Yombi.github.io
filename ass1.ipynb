```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def F(z,t):
    x,y=z
    dxdt=y
    dydt=6*x-5*y
    return np.array([dxdt,dydt])
z0=np.array([2,3])
t=np.linspace(0,1,100)
z=odeint(F,z0,t)
x=z[:,0]
y=z[:,1]
z1=(-1/7)*np.exp(-6*t)+(15/7)*np.exp(t)
plt.subplot(1,2,1)
plt.plot(t,x,"r--",linewidth=10,label="num_sol")
#plt.plot(t,y)
plt.plot(t,z1,"b:",linewidth=5,label="exact_sol")
plt.legend()
plt.subplot(1,2,2)
plt.plot(t,abs(x-z1))
plt.show()
```
